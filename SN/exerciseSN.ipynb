```python
import numpy as np
import matplotlib.pyplot as plt
```

leggo il file dei dati


```python
data = np.loadtxt("fit_exercise_data.txt")
#separaz colonne , colonna 1 =redshift z, colonna 2 modulo distanza mu
z = data[:, 0]
mu = data[:, 1]
```

plot del grafico

```python
plt.figure(figsize=(8, 6))
plt.scatter(z, mu, color='b', label="Dati Supernovae", s=10, alpha=0.7)  # Punti dati
plt.xlabel("Redshift (z)")
plt.ylabel("Modulo della distanza mu (mag)")
plt.title("Supernovae Ia: Modulo della distanza vs Redshift")
plt.xscale("log")  # Scala logaritmica per il redshift per una migliore visualizzazione
plt.legend()
#plt.grid(True, which="both", linestyle="--", alpha=0.5)

# Mostra il plot
plt.show()
```

Ora aggiungiamo errori usando la matrice di covarianza fornita dal file

```python
#carico il file dei dati covarianza
cov_matrix = np.loadtxt("fit_exercise_covariance.txt")
```

```python
#calcola la barra di errore
errors = np.sqrt(np.diag(cov_matrix))
```

```python
#ri-plotto con errori

plt.figure(figsize=(8, 5))
plt.errorbar( z, mu, yerr=errors, fmt='o', capsize=5, label="Dati con errori", color="blue")
plt.xscale("log")  # Scala logaritmica per il redshift per una migliore visualizzazione
# Personalizzazioni
plt.xlabel("Indice dei dati")
plt.ylabel("Valore misurato")
plt.title("Dati con barre d'errore")
plt.legend()
#plt.grid(True)

# Mostrare il grafico
plt.show()
```
