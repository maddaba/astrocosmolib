```python
print(data.columns.names)
```

```python
ra=data["alpha"] #ascensione retta
dec=data["delta"]#declinazione
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.scatter(ra,dec,s=1, color='blue',alpha=0.5)
plt.xlabel("Ascensione retta (RA)")
plt.ylabel("Declinazione (DEC)")
plt.title("Distrib degli oggetti nel cielo")
plt.grid(True)
plt.show()
```

```python
zsp=data["zspec"]
```

```python
plt.hist(zsp,bins=100)
plt.xlabel("zspec")
plt.ylabel("Distrib del redshift")
plt.show()
```
